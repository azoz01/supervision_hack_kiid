```python
from lib.constants import RAW_KIIDS_DIR, META_FILE_PATH
```

```python
list_of_files = RAW_KIIDS_DIR.iterdir()
list_of_files = [f.name for f in list_of_files if f.suffix in [".ashx", ".pdf"]]
```

```python
import pandas as pd
out_df = pd.DataFrame(columns=["ID_KIID", "ID_ZESPOLU", "NAZWA_PLIKU"])
id = 0
for file in list_of_files:
    id += 1
    row_dict = {
        "ID_KIID": id,
        "ID_ZESPOLU": 2,
        "NAZWA_PLIKU": file
    }
    row_df = pd.DataFrame(data=row_dict, index=[0])
    out_df = pd.concat([out_df, row_df])
out_df.to_csv(META_FILE_PATH, index=False)
```

```python
import pandas as pd
from lib.get_fund_info import retive_all_info
from lib.constants import PARSED_KIIDS_DIR
```

```python
df = pd.read_parquet(PARSED_KIIDS_DIR)
knf_df = pd.read_pickle("data/knf_funds.pkl")
```

```python
from collections import Counter
```

```python
def calculate_bow_counter_for_record(record_data):
    intro_text = record_data["intro"]
    fund_info = retive_all_info(knf_df, intro_text)
    name = fund_info["subfund_name"] if fund_info["subfund_name"] else fund_info["fund_name"]
    text = record_data["raw_text"].lower()
    splitted = text.split()
    counter = Counter(splitted)
    return name, counter
```

```python
funds_dict = {}
for index, record_data in df.iterrows():
    name, counter = calculate_bow_counter_for_record(record_data)
    funds_dict[name] = funds_dict.get(name, Counter()) + counter
```
